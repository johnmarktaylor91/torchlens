TorchLens bundle diff hero: clean ResNet-18 vs zero-ablated `layer1.0.relu`.

```python
import warnings

import torch
from torchvision.models import resnet18

import torchlens as tl
from torchlens.intervention import MultiMatchWarning

warnings.filterwarnings("ignore", category=MultiMatchWarning)

torch.manual_seed(0)
model = resnet18(weights=None).eval()
x = torch.randn(1, 3, 224, 224)

model_log = tl.log_forward_pass(model, x, vis_opt="none", intervention_ready=True)
ablated_log = model_log.fork("ablated")
ablated_log.do(tl.module("layer1.0.relu"), tl.zero_ablate(), confirm_mutation=True)
bundle = tl.bundle({"clean": model_log, "ablated": ablated_log}, baseline="clean")

bundle.show_diff(
    vis_outpath="examples/demos/bundle_diff_clean_vs_zero_relu",
    vis_save_only=True,
)
```
