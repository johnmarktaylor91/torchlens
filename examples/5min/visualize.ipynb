# Visualize a model graph

Graph views help you map layer names to model operations before extracting or intervening.

```python
from pathlib import Path
import importlib.util
import sys

from IPython.display import SVG, display
import torch
import torchlens as tl
from torchlens.visualization import show_model_graph

shared_path = None
for root in [Path.cwd(), *Path.cwd().parents]:
    candidate = root / "notebooks" / "total_audit" / "_shared.py"
    if candidate.exists():
        shared_path = candidate
        break
if shared_path is None:
    raise FileNotFoundError("Could not find notebooks/total_audit/_shared.py")

spec = importlib.util.spec_from_file_location("torchlens_total_audit_shared", shared_path)
if spec is None or spec.loader is None:
    raise ImportError(f"Could not load {shared_path}")
shared = importlib.util.module_from_spec(spec)
sys.modules[spec.name] = shared
spec.loader.exec_module(shared)
tiny_model = shared.tiny_model

torch.manual_seed(0)
model = tiny_model(seed=7).eval()
x = torch.randn(2, 4)
```

This is the three-line version: pick an output path, render the graph, and display the SVG.

```python
graph_path = Path("/tmp/torchlens_5min_visualize.svg")
show_model_graph(
    model, x, vis_outpath=str(graph_path.with_suffix("")), vis_fileformat="svg", vis_save_only=True
)
display(SVG(filename=str(graph_path)))
```

The graph labels are the same labels used by `peek`, `extract`, and intervention selectors.
