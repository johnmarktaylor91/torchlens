# Find your first NaN

When a model emits NaNs or infinities, TorchLens can tell you which operation produced the first non-finite tensor.

```python
from __future__ import annotations

import torch
from torch import nn
import torchlens as tl


class NonFiniteFixture(nn.Module):
    """Tiny model that intentionally creates a NaN/Inf activation."""

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """Divide by zero after one harmless affine-style step."""

        hidden = x + 1.0
        return hidden / (hidden - hidden)


torch.manual_seed(0)
model = NonFiniteFixture().eval()
x = torch.ones(1, 2)
```

The user path is two lines: log the forward pass, then ask for the first non-finite layer.

```python
log = tl.log_forward_pass(model, x)
print(log.first_nonfinite())
```

`first_nonfinite()` is useful in notebooks because it returns a readable diagnostic instead of forcing an exception-driven workflow.
